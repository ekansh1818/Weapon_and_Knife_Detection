```python
plt.figure(figsize=(12, 8))

# Plotting with different line styles, markers, and colors
plt.plot(df[["train/cls_loss"]], label="No model", linestyle='-', color="purple")
plt.plot(df_db[["train/cls_loss"]], label="Daubechies", linestyle='-', color="blue")
plt.plot(df_haar[["train/cls_loss"]], label="Haar", linestyle='-', color="green")
plt.plot(df_sym[["train/cls_loss"]], label="Symlet", linestyle='-', color="red")
plt.plot(df_compress[["train/cls_loss"]], label="No model Compressed", linestyle='-', color="orange")
plt.plot(df_haar_compress[["train/cls_loss"]], label="Haar Compressed", linestyle='-', color="yellow")

# Adding grid and adjusting legend
plt.grid(True)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Categorical Cross-Entropy Loss")
plt.legend(loc="upper right")

plt.show()
```
